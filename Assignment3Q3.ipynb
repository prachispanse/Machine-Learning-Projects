```python
import numpy as np 
from sklearn.tree import DecisionTreeRegressor
import random
import sklearn.metrics as metrics
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

def import_data(filename, delimiter):
    training_data = list()
    with open(filename) as tr:
        read_data = tr.read()
        for row in read_data.split("\n"):
            training_data.append(row.split(delimiter))
    training_data = training_data[:len(training_data)-1]
    # print(str(len(training_data)))
    return training_data

def get_random_indices(max, count):
    indices = set()
    counter = 0
    while counter<count:
        index = random.randint(0, max)
        if index not in indices:
            indices.add(index)
            counter += 1
    return indices

def filter(data, count):
    filtered_data = list()
    for i in range(0,10):
        filtered_data.append(list())
    for i in range(0, len(data)):
        filtered_data[int(data[i][len(data[i])-1])].append(data[i][:len(data[i])-1])
    filtered_counts_of_data = list()
    filtered_data_classes = list()
    for i in range(0,10):
        filtered_counts_of_data.extend(filtered_data[i][:count])
        for j in range(0,count):
            filtered_data_classes.append(i)
    return filtered_counts_of_data, filtered_data_classes
        
def filter_6_and_9(training_data, count):
#     indices = get_random_indices(len(training_data), counter)
    training_data_class_6_and_9_48 = list()
    training_data_classes_48 = list()
    training_data_class_6_and_9_64 = list()
    training_data_classes_64 = list()
    for i in range(0, len(training_data)):
        if(training_data[i][len(training_data[i])-1] == '6'):
            training_data_class_6_and_9_48.append(training_data[i][:len(training_data[i])-17])
            training_data_classes_48.append(training_data[i][len(training_data[i])-1])
            training_data_class_6_and_9_64.append(training_data[i][:len(training_data[i])-1])
            training_data_classes_64.append(training_data[i][len(training_data[i])-1])
        elif (training_data[i][len(training_data[i])-1] == '9'):
            training_data_class_6_and_9_48.append(training_data[i][:len(training_data[i])-17])
            training_data_classes_48.append(training_data[i][len(training_data[i])-1])
            training_data_class_6_and_9_64.append(training_data[i][:len(training_data[i])-1])
            training_data_classes_64.append(training_data[i][len(training_data[i])-1])
    return (training_data_class_6_and_9_48, training_data_classes_48, training_data_class_6_and_9_64, training_data_classes_64 )

def get_random_class_6_and_9(data_class_6_and_9, count):
    data_class6, data_class9, data_class_6_and_9, data_classes = filter_6_and_9(data_class_6_and_9)
    indices_for_class6 = get_random_indices(len(data_class6))
    indices_for_class9 = get_random_indices(len(data_class9))
    randomized_data_class_6_and_9 = list()
    randomized_data_classes = list()
    for i in indices_for_class6:
        if i < len(data_class6):
            randomized_data_class_6_and_9.append(data_class6[i])
            randomized_data_classes.append(6)
#     print ("indices_for_class6 :" + str(len(indices_for_class6)))
#     print ("randomized_data_class_6_and_9 :" + str(len(randomized_data_class_6_and_9)))
    for i in indices_for_class9:
        if i < len(data_class9):
            randomized_data_class_6_and_9.append(data_class9[i])
            randomized_data_classes.append(9)
#     print ("indices_for_class9 :" + str(len(indices_for_class9)))
#     print ("randomized_data_class_6_and_9 :" + str(len(randomized_data_class_6_and_9)))
    return (randomized_data_class_6_and_9, randomized_data_classes)

def filter_column(arr, col):
    feature = list()
    for i in range(0, len(arr)):
        feature.append(arr[i][col])
    return feature

def find_total_error(data_classes, predicted):
    error = 0
#     print("data_classes : " + str(data_classes))
#     print("predicted : " + str(predicted))
    predicted_list = predicted.tolist()
    for i in range(0, len(data_classes)):
        if float(data_classes[i]) != float(predicted_list[i]):
           error += 1
    return error

def feature_regressor(training_data_class_6_and_9_500_48, training_data_class_6_and_9_500_64, testing_data_class_6_and_9_500_48, testing_data_class_6_and_9_500_64):
    for i in range(48, 64):
        y_train=filter_column(training_data_class_6_and_9_500_64, i)
        y_test = filter_column(testing_data_class_6_and_9_500_64, i)
        regressor = DecisionTreeRegressor(random_state=0)
        regressor.fit(training_data_class_6_and_9_500_48, y_train)
        predicted = regressor.predict(testing_data_class_6_and_9_500_48)
        maerror = metrics.mean_absolute_error(np.array(y_test).astype(int), np.array(predicted).astype(int))
#         print("maerror for feature " + str(i+1) + " : " + str(maerror))
        mserror = metrics.mean_squared_error(np.array(y_test).astype(int), np.array(predicted).astype(int))
        print("mserror for feature " + str(i+1) + " : " + str(mserror))
    
    
training_data = import_data('optdigits.tra', "\t")
testing_data = import_data('optdigits.tes', "\t")


print("\nX500_69")
training_data.extend(training_data)
testing_data.extend(testing_data)
training_data_class_6_and_9_500_48, training_data_classes_500_48, training_data_class_6_and_9_500_64, training_data_classes_500_64 = filter_6_and_9(training_data, 500)
testing_data_class_6_and_9_500_48, testing_data_classes_500_48, testing_data_class_6_and_9_500_64, testing_data_classes_500_64 = filter_6_and_9(testing_data, 500)
feature_regressor(training_data_class_6_and_9_500_48, training_data_class_6_and_9_500_64, testing_data_class_6_and_9_500_48, testing_data_class_6_and_9_500_64)

```
