```python
import numpy as np 
from sklearn.neighbors import KNeighborsClassifier
import random
from sklearn.metrics import accuracy_score
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

def import_data(filename, delimiter):
    training_data = list()
    with open(filename) as tr:
        read_data = tr.read()
        for row in read_data.split("\n"):
            training_data.append(row.split(delimiter))
    training_data = training_data[:len(training_data)-1]
    # print(str(len(training_data)))
    return training_data

def get_random_indices(max, count):
    indices = set()
    counter = 0
    while counter<count:
        index = random.randint(0, max)
        if index not in indices:
            indices.add(index)
            counter += 1
    return indices

def filter(data, count):
    filtered_data = list()
    for i in range(0,10):
        filtered_data.append(list())
    for i in range(0, len(data)):
        filtered_data[int(data[i][len(data[i])-1])].append(data[i][:len(data[i])-1])
    filtered_counts_of_data = list()
    filtered_data_classes = list()
    for i in range(0,10):
        filtered_counts_of_data.extend(filtered_data[i][:count])
        for j in range(0,count):
            filtered_data_classes.append(i)
    return filtered_counts_of_data, filtered_data_classes
        
def filter_6_and_9(training_data, count):
#     indices = get_random_indices(len(training_data), counter)
    training_data_class_6_and_9 = list()
    training_data_class6 = list()
    training_data_class9 = list()
    training_data_classes = list()
    training_counter6=0
    training_counter9=0
    for i in range(0, len(training_data)):
        if(training_data[i][len(training_data[i])-1] == '6') and training_counter6<count:
            training_data_class_6_and_9.append(training_data[i][:len(training_data[i])-1])
            training_data_classes.append(training_data[i][len(training_data[i])-1])
            training_data_class6.append(training_data[i][:len(training_data[i])-1])
            training_counter6 +=1
        elif (training_data[i][len(training_data[i])-1] == '9') and training_counter9<count:
            training_data_class_6_and_9.append(training_data[i][:len(training_data[i])-1])
            training_data_classes.append(training_data[i][len(training_data[i])-1])
            training_data_class9.append(training_data[i][:len(training_data[i])-1])
            training_counter9 +=1
    return (training_data_class6, training_data_class9, training_data_class_6_and_9, training_data_classes )

def get_random_class_6_and_9(data_class_6_and_9, count):
    data_class6, data_class9, data_class_6_and_9, data_classes = filter_6_and_9(data_class_6_and_9)
    indices_for_class6 = get_random_indices(len(data_class6))
    indices_for_class9 = get_random_indices(len(data_class9))
    randomized_data_class_6_and_9 = list()
    randomized_data_classes = list()
    for i in indices_for_class6:
        if i < len(data_class6):
            randomized_data_class_6_and_9.append(data_class6[i])
            randomized_data_classes.append(6)
#     print ("indices_for_class6 :" + str(len(indices_for_class6)))
#     print ("randomized_data_class_6_and_9 :" + str(len(randomized_data_class_6_and_9)))
    for i in indices_for_class9:
        if i < len(data_class9):
            randomized_data_class_6_and_9.append(data_class9[i])
            randomized_data_classes.append(9)
#     print ("indices_for_class9 :" + str(len(indices_for_class9)))
#     print ("randomized_data_class_6_and_9 :" + str(len(randomized_data_class_6_and_9)))
    return (randomized_data_class_6_and_9, randomized_data_classes)
            

def find_total_error(data_classes, predicted):
    error = 0
#     print("data_classes : " + str(data_classes))
#     print("predicted : " + str(predicted))
    predicted_list = predicted.tolist()
    for i in range(0, len(data_classes)):
        if data_classes[i] != predicted_list[i]:
           error += 1
    return error

def kparameter (depth, training_data_class_6_and_9, training_data_classes, testing_data_class_6_and_9, testing_data_classes):
    neigh =  KNeighborsClassifier(n_neighbors = depth)
    neigh.fit(training_data_class_6_and_9, training_data_classes)
    training_predicted = neigh.predict(training_data_class_6_and_9)
#     print("training_predicted : " + str(training_predicted))
#     print("training_actual : " + str(training_data_classes))
    training_error = find_total_error(training_data_classes, training_predicted)
#     print("training_error : " + str(training_error))
    testing_predicted = neigh.predict(testing_data_class_6_and_9)
#     print("testing_predicted : " + str(testing_predicted))
#     print("testing_actual : " + str(testing_data_classes))
    testing_error = find_total_error(testing_data_classes, testing_predicted)
    acc_s=accuracy_score(testing_data_classes, testing_predicted)
    print("k_value :" + str(depth) + "\ttesting_error : " + str("{:.5f}".format(1-acc_s)) + "\taccuracy_score : " + str("{:.5f}".format(acc_s)))
    
training_data = import_data('optdigits.tra', "\t")
training_class6, training_class9, training_data_class_6_and_9, training_data_classes = filter_6_and_9(training_data, 100)
# print ("training_data_class_6_and_9:" + str(len(training_data_class_6_and_9)))
testing_data = import_data('optdigits.tes', "\t")
testing_class6, testing_class9, testing_data_class_6_and_9, testing_data_classes = filter_6_and_9(testing_data, 100)
# print ("testing_data_class_6_and_9:" + str(len(testing_data_class_6_and_9)))
filtered_multi_class_training_data, multi_class_data_classes = filter(training_data, 100)
filtered_multi_class_testing_data, all_classes = filter(testing_data, 100)
# print ("filtered_multi_class_training_data:" + str(len(filtered_multi_class_training_data)))
# print ("filtered_multi_class_testing_data:" + str(len(filtered_multi_class_testing_data)))

print("\nX100_C69")
kparameter(1, training_data_class_6_and_9, training_data_classes, testing_data_class_6_and_9, testing_data_classes)
kparameter(3, training_data_class_6_and_9, training_data_classes, testing_data_class_6_and_9, testing_data_classes)
kparameter(5, training_data_class_6_and_9, training_data_classes, testing_data_class_6_and_9, testing_data_classes)
kparameter(9, training_data_class_6_and_9, training_data_classes, testing_data_class_6_and_9, testing_data_classes)

print("\nX100_CAll")
kparameter(1, filtered_multi_class_training_data, multi_class_data_classes, filtered_multi_class_testing_data, all_classes)
kparameter(3, filtered_multi_class_training_data, multi_class_data_classes, filtered_multi_class_testing_data, all_classes)
kparameter(5, filtered_multi_class_training_data, multi_class_data_classes, filtered_multi_class_testing_data, all_classes)
kparameter(9, filtered_multi_class_training_data, multi_class_data_classes, filtered_multi_class_testing_data, all_classes)

print("\nX500_69")
training_data.extend(training_data)
testing_data.extend(testing_data)
testing_data.extend(testing_data)
training_class6_500, training_class9_500, training_data_class_6_and_9_500, training_data_classes_500 = filter_6_and_9(training_data, 500)
testing_class6_500, testing_class9_500, testing_data_class_6_and_9_500, testing_data_classes_500 = filter_6_and_9(testing_data, 500)
kparameter(1, training_data_class_6_and_9_500, training_data_classes_500, testing_data_class_6_and_9_500, testing_data_classes_500)
kparameter(3, training_data_class_6_and_9_500, training_data_classes_500, testing_data_class_6_and_9_500, testing_data_classes_500)
kparameter(5, training_data_class_6_and_9_500, training_data_classes_500, testing_data_class_6_and_9_500, testing_data_classes_500)
kparameter(9, training_data_class_6_and_9_500, training_data_classes_500, testing_data_class_6_and_9_500, testing_data_classes_500)

print("\nX500_CAll")
filtered_multi_class_training_data_500, multi_class_data_classes_500 = filter(training_data, 500)
filtered_multi_class_testing_data_500, classes_500 = filter(testing_data, 500)
kparameter(1, filtered_multi_class_training_data_500, multi_class_data_classes_500, filtered_multi_class_testing_data_500,classes_500)
kparameter(3, filtered_multi_class_training_data_500, multi_class_data_classes_500, filtered_multi_class_testing_data_500, classes_500)
kparameter(5, filtered_multi_class_training_data_500, multi_class_data_classes_500, filtered_multi_class_testing_data_500, classes_500)
kparameter(9, filtered_multi_class_training_data_500, multi_class_data_classes_500, filtered_multi_class_testing_data_500, classes_500)

```
